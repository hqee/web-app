## 1. Basic Line Plot

```python
# kita membuat basic line plot
import matplotlib.pyplot as plt
import numpy as np

myarray = np.array([1, 2, 3])
plt.plot(myarray)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

### Penjelasan Kode

Block ini membuat line plot sederhana menggunakan matplotlib. Data yang dipakai berupa array myarray = [1, 2, 3]. Fungsi plt.plot() menggambar garis berdasarkan data tersebut. Kemudian plt.xlabel() dan plt.ylabel() memberi label pada sumbu x dan y. Terakhir, plt.show() menampilkan grafik ke layar.

### Penjelasan Library

- numpy (np) → digunakan untuk membuat array numerik (np.array([1, 2, 3])) agar data bisa diproses dengan mudah.

- matplotlib.pyplot (plt) → digunakan untuk membuat grafik (line plot) dan menambahkan label pada sumbu.
___

### 2. Scatter Plot

```python
# membuat scatter plot
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])

plt.scatter(x, y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

## Penjelasan Kode

Block ini membuat scatter plot atau grafik titik. Dua array didefinisikan, x = [1, 2, 3] dan y = [4, 5, 6]. Fungsi plt.scatter(x, y) menggambar titik-titik berdasarkan pasangan nilai (x, y). Sama seperti sebelumnya, sumbu x dan y diberi label, lalu grafik ditampilkan dengan plt.show().

## Penjelasan Library

- numpy (np) → digunakan untuk membuat array numerik (x dan y).

- matplotlib.pyplot (plt) → digunakan untuk membuat grafik scatter, menambahkan label, dan menampilkan hasilnya.
