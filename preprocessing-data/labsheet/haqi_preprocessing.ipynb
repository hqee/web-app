# Pembagian Dataset

```python
# import library dan fungsi
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
filename = 'heart_disease.csv' # memasukkan nama file 'heart_disease.csv' ke dalam variabel filename
# membuat sebuah list yang berisi nama-nama kolom
names = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'smoking','diabetes', 'bmi', 'heart_disease']
df = pd.read_csv(filename, names=names, header=None, skiprows=1) # Membaca file CSV dengan beberapa aturan khusus
df.head() # menampilkan 5 baris pertama dari dataframe df
```

```python
array = df.values # mengubah dataframe df menjadi array
X = array[:, 0:8] # Memisahkan fitur (kolom 0 hingga 7) ke dalam variabel X
y = array[:, 8] # Memisahkan target/label (kolom 8) ke dalam variabel y
test_size = 0.33 # proporsi data uji sebesar 33%
seed = 7 # nilai seed untuk random state agar hasil pembagian data bisa direproduksi

# Pembagian Dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=seed)

logreg = LogisticRegression(max_iter=1000) # Membuat model regresi logistik dengan batas maksimum iterasi 1000
logreg.fit(X_train, y_train) # Melatih model regresi logistik menggunakan data latih (X_train dan y_train)
result = logreg.score(X_test, y_test) # Menghitung akurasi model pada data uji (X_test dan y_test)

# menampilkan akurasi model dalam persentase
print("Accuracy:", result*100) 
```

```python
from sklearn.model_selection import KFold # import KFold untuk cross validation
from sklearn.model_selection import cross_val_score # import cross_val_score untuk cross validation

num_folds = 10 # jumlah fold sebanyak 10
```

```python
kfold = KFold(n_splits=num_folds) # Membuat objek KFold untuk melakukan cross-validation dengan jumlah fold yang ditentukan
result_cv = cross_val_score(logreg, X, y, cv=kfold) # Melakukan cross-validation pada model logreg menggunakan data X dan y dengan pembagian kfold

# menampilkan rata-rata akurasi dan standar deviasi dari hasil cross-validation dalam persentase
print("Accuracy:", result_cv.mean()*100) 
print("Accuracy:", result_cv.std()*100)
```

```python
# evaluate menggunakan Leave One Out Cross Validation
from sklearn.model_selection import LeaveOneOut # import LeaveOneOut untuk cross validation

loocv = LeaveOneOut() # Membuat objek LeaveOneOut untuk melakukan cross-validation dengan metode Leave-One-Out
result_loocv = cross_val_score(logreg, X, y, cv=loocv) # Melakukan cross-validation pada model logreg menggunakan data X dan y dengan metode Leave-One-Out
```

```python
# menampilkan rata-rata akurasi dan standar deviasi dari hasil Leave-One-Out Cross-Validation dalam persentase
print("Accuracy:", result_loocv.mean()*100) 
print("Accuracy:", result_loocv.std()*100)
```

```python
# shuffle split
from sklearn.model_selection import ShuffleSplit # import ShuffleSplit untuk cross validation

ssplit = ShuffleSplit(n_splits=num_folds, test_size=test_size, random_state=seed) # Membuat objek ShuffleSplit untuk melakukan cross-validation dengan metode shuffle split
result_ssplit = cross_val_score(logreg, X, y, cv=ssplit) # Melakukan cross-validation pada model logreg menggunakan data X dan y dengan metode Shuffle Split
```

```python
# menampilkan rata-rata akurasi dan standar deviasi dari hasil Shuffle Split dalam persentase
print("Accuracy:", result_ssplit.mean()*100)
print("Accuracy:", result_ssplit.std()*100)
```

# Transformasi Data

```python
# import library dan fungsi
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
filename = 'heart_disease.csv' # memasukkan nama file 'heart_disease.csv' ke dalam variabel filename
# membuat sebuah list Python yang berisi nama-nama kolom
names = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'smoking','diabetes', 'bmi', 'heart_disease']
df = pd.read_csv(filename, names=names, header=None, skiprows=1) # membaca file CSV dengan nama kolom sesuai list 'names', tanpa membaca header asli file, dan melewati baris pertama (header)
df.head() # menampilkan 5 baris pertama dari dataframe df
```

```python
array = df.values # mengubah dataframe df menjadi array
print(type(array)) # menampilkan tipe data array
print(array) # menampilkan isi array
```

```python
X = array[:, 0:8] # Memisahkan fitur (kolom 0 hingga 7) ke dalam variabel X
y = array[:, 8] # Memisahkan target/label (kolom 8) ke dalam variabel y
print(X.shape) # menampilkan dimensi dari X
print(y.shape) # menampilkan dimensi dari y
```

## Rescale Data

```python
# rescale data
scaler = MinMaxScaler(feature_range=(0, 1)) # Membuat objek MinMaxScaler untuk merescaling fitur ke dalam rentang 0 hingga 1
rescaledX = scaler.fit_transform(X) # Melakukan fit dan transform pada data X untuk merescaling setiap fitur ke dalam rentang yang ditentukan

# set print options dengan numpy
np.set_printoptions(precision=3) # mengatur numpy untuk menampilkan angka dengan presisi 3 desimal

# cetak data setelah rescale 5 baris pertama tiap column X
print(rescaledX[0:5, :]) # Menampilkan 5 baris pertama dari data yang telah direscale
```

## Standardize Data

```python
from sklearn.preprocessing import StandardScaler # import StandardScaler dari library sklearn

scaler2 = StandardScaler().fit(X) # Membuat objek StandardScaler dan melakukan fit pada data X untuk menghitung mean dan standar deviasi dari setiap fitur
rescaledX2 = scaler2.transform(X) # Melakukan transform pada data X untuk menstandarisasi setiap fitur sehingga memiliki mean 0 dan standar deviasi 1

# set print options dengan numpy
np.set_printoptions(precision=3) # mengatur numpy untuk menampilkan angka dengan presisi 3 desimal

# cetak data setelah standardize 5 baris pertama tiap column X
print(rescaledX2[0:5, :]) 
```

## Normalize Data

```python
from sklearn.preprocessing import Normalizer # import Normalizer dari library sklearn

scaler3 = Normalizer().fit(X) # Membuat objek Normalizer dan melakukan fit pada data X
normalizedX = scaler3.transform(X) # Melakukan transform pada data X untuk menormalkan setiap fitur sehingga memiliki norma (panjang vektor) 1

# set print options dengan numpy
np.set_printoptions(precision=3) # mengatur numpy untuk menampilkan angka dengan presisi 3 desimal

# cetak data setelah normalize 5 baris pertama tiap column X
print(normalizedX[0:5, :])
```

## Binarize Data

```python
from sklearn.preprocessing import Binarizer # import Binarizer dari library sklearn

binarizer = Binarizer(threshold=0.0).fit(X) # Membuat objek Binarizer dengan threshold 0.0 dan melakukan fit pada data X
binaryX = binarizer.transform(X) # Melakukan transform pada data X untuk mengubah setiap nilai fitur menjadi 0 atau 1 berdasarkan threshold yang ditentukan

# set print options dengan numpy
np.set_printoptions(precision=3) # mengatur numpy untuk menampilkan angka dengan presisi 3 desimal

# cetak data setelah binarize 5 baris pertama tiap column X
print(binaryX[0:5, :])
```
