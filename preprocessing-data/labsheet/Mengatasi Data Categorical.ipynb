```python
import pandas as pd
```

```python
netflix = pd.read_csv('Netflix.csv')
netflix.head()
```

```python
netflix.info()
```

```python
data_cat = netflix['type']
data_cat.value_counts()
```

# 1. Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
data_cat_encoded = label_encoder.fit_transform(data_cat.values)
data_cat_encoded
```

```python
print(label_encoder.classes_)
```

# 2. One Hot Encoder

```python
from sklearn.preprocessing import OneHotEncoder
one_hot = OneHotEncoder()
data_cat_1hot = one_hot.fit_transform(data_cat.values.reshape(-1, 1))
data_cat_1hot.toarray()
```

# 3. LabelBinarizer

```python
from sklearn.preprocessing import LabelBinarizer
binarizer = LabelBinarizer()
data_cat_binarizer = binarizer.fit_transform(data_cat)
data_cat_binarizer
```
