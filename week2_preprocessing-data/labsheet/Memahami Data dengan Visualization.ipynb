# 1. Univariate Plots

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filename = 'diabetes.csv'
names = ['preg', 'gluc', 'blood', 'skin', 'ins', 'bmi', 'ped', 'age', 'class']
data = pd.read_csv(filename, names=names, header=None, skiprows=1)
data.head()
```

```python
# histogram
data.hist(figsize=(12, 12))
```

```python
# density plots
data.plot(kind='density', subplots=True, layout=(3,3), sharex=False, figsize=(12,12))
```

```python
# box and whisker plots
data.plot(kind='box', subplots=True, layout=(3,3), sharex=False, sharey=False, figsize=(12, 12))
```

# 2. Multivariate Plots

```python
import numpy as np
# correlation matrix plots
correlations = data.corr()

# plot correlation
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,9,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
plt.show()
```
